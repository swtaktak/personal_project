## 구매 리텐션

구매 리텐션을 볼려고 한다. 롤링도 의미가 있겠으나, 월이라는 기간을 잡고 리텐션 보자.

```python
df_online_sales
```

```python
df_buy_history = df_online_sales[['고객ID', '거래날짜']].drop_duplicates()
df_buy_history['거래월'] = df_online_sales['거래날짜'].apply(lambda x: x[0:7])
```

```python
df_buy_history = pd.merge(df_buy_history, df_user_first, on = '고객ID', how = 'left')
```

```python
def get_month_diff(now, first):
    now_month_str = now[-2:]
    first_month_str = first[-2:]
    now_month = int(now_month_str) if now_month_str[0] == '1' else int(now_month_str[-1])
    first_month = int(first_month_str) if first_month_str[0] == '1' else int(first_month_str[-1])
    return now_month - first_month
```

```python
df_buy_history['경과월'] = df_buy_history.apply(lambda x : get_month_diff(x['거래월'], x['최초구매월']), axis = 1)
```

```python
df_buy_history
```

```python
df_ret_base = df_buy_history[['고객ID', '최초구매월', '경과월']].drop_duplicates()
```

```python
df_ret_base
```

```python
df_ret = df_ret_base.groupby(['최초구매월', '경과월']).count().reset_index(drop = False)
df_ret.columns = ['최초구매월', '경과월', '구매리텐션_유저수']
df_ret_pivot = df_ret.pivot(index = '최초구매월', columns = '경과월', values = '구매리텐션_유저수')
```

```python
df_ret_pivot
```

```python
plt.figure(figsize = (12, 10))
sns.heatmap(df_ret_pivot, annot = True)
plt.show()
```

정확한 비율에 대한 확인이 필요하겠으나, 각 최초 구매일 유저 집단별로, 리텐션이 좋지 아니하다

## 리텐션 비율 검증

```python
df_ret_max = df_ret.loc[df_ret['경과월']==0].drop('경과월', axis = 1)
df_ret_max.columns = ['최초구매월', '최대리텐션']
df_ret = pd.merge(df_ret, df_ret_max, on = '최초구매월', how = 'left')
df_ret['리텐션비율'] = round(df_ret['구매리텐션_유저수']/df_ret['최대리텐션'], 2)
df_ret
```

```python
df_ret_ratio_pivot = df_ret.pivot(index = '최초구매월', columns = '경과월', values = '리텐션비율')
plt.figure(figsize = (12, 10))
sns.heatmap(df_ret_ratio_pivot, annot = True)
plt.show()
```

롤링 리텐션으로 추가 확인을 해볼 수도 있겠으나, 이 정보 만으로도 재구매율이 낮다는 것을 알 수 있어, 유저들의 재구매율을 높이기 위한 방안이 필요하다

## 가설 설정

현재 쇼핑몰의 문제는 바로 낮은 재구매율이다. 낮은 재구매율을 높이기 위해 현재의 쿠폰 시스템이 잘 이뤄지고 있는지 확인해 봐야 할 것이다.

### 재구매의 정의

하루, 이틀 차이가 아니라, 어느정도 서비스를 유의미하게 기억하고 있는가를 기준으로 할 것이다. 따라서 서로 다른 2개 이상의 월에 구매를 한 경험이 있다면 재구매를 했다고 잡아보려고 한다. 이를 바탕으로 가설을 세워보자.

가설 : 쿠폰 사용 경험이 있는 유저들은 그렇지 않은 유저들보다 재구매할 가능성이 높다.

단순히 쿠폰을 클릭해보았다고 사용으로 이어졌다고 보기는 어려우므로 실제 사용 경험을 기준으로 쿠폰 사용 여부를 나누는 것이 적합할 것이다.

```python
df_coupon = df_online_sales[['고객ID', '쿠폰상태']].drop_duplicates()
df_coupon['쿠폰사용여부'] = df_coupon['쿠폰상태'].apply(lambda x: 1 if x == 'Used' else 0)
df_coupon = df_coupon.sort_values('쿠폰사용여부', ascending = False).groupby('고객ID').head(1)
df_coupon = df_coupon.drop('쿠폰상태', axis = 1)
```

```python
df_user_info = pd.merge(df_user_info, df_coupon, on = '고객ID', how='left')
```

```python
df_user_info['재방문여부'] = df_user_info.apply(lambda x: 1 if x['최초구매월'] != x['최종구매월']
                                          else 0, axis = 1)
```

```python
pd.crosstab(index = df_user_info['재방문여부'], columns = df_user_info['쿠폰사용여부'])
```

두 가지 해석이 가능한 상황이다.

- 쿠폰 사용 여부가 재방문율에 유의미한 영향을 주었는가?
  * 다시 말해, 쿠폰 사용 경험이 있는 유저는 재방문을 했을까?
    - Problem : 쿠폰을 사용하기 위해 재방문을 했을수도 있다라는 문제점이 존재한다. (하지만 아무튼 매출은 늘 것이므로 상관없다.)
- 재방문을 한 유저는 쿠폰을 많이 사용했을까?

쿠폰 -> 재방문의 검증 결과는
  - 쿠폰 미사용 유저 : 8/(8+86) = 8.5%
  - 쿠폰 사용 유저 : 572/(572+802) = 41.6%

통계 검정도 갈 것 없이, 명확하게 쿠폰을 사용한 유저의 재방문율이 더 높다는 것을 알 수 있다.

재방문-> 쿠폰의 검증 결과는
- 재방문 경험이 없는 유저 : 802/(86+802) = 90.3%
- 재방문 경험이 있는 유저 : 572/(572+8) = 98.6%

역시 재방문 경험이 있는 유저들이 쿠폰을 더 많이 썼다. 재방문 경험이 없어도 쿠폰을 사용한 유저들이 잇는 것으로 보면 쿠폰의 발매 시점 문제는 아니고, 쿠폰 사용이 유의미한 긍정적 경험을 가져올 수 있음을 알 수 있다. 따라서 쿠폰을 사용하는 마케팅을 고려하자.

## 어떤 유저들에게 쿠폰 캠페인 마케팅을 실시해야 할까?

아쉽게도 마케팅 비용에 대한 정보가 추가로 주어져 있지 않아 필요한 전환율 추정은 쉽지 않다. 마케팅 방식에 대한 정보도 주어져 있지 아니하므로, 다른 방법을 생각해야 할 것이다.

쿠폰의 사용 경험이 없는 94명의 유저들에게 좋은 변화를 만들 수 있을까?
