### 근로기간

```python
train_df['근로기간'].value_counts()
```

전처리가 필요합니다. 우선 통합되지 않고 같은 의미이나 자기멋대로 분리된 것들은 합쳐줍시다.

또한, 근로 기간을 모를 경우에는 단순히 unknown으로 두기에는 좀 그렇습니다.

* 데이터가 알려지지 않은 경우일수도 있지만, 본인의 근로기간을 모른다는 것이 말이 안되므로 0이 합리적인 판단일겁니다.
* 10+일 경우 역시 그 이상에 대한 판단이 불가능하여 10으로 통일합니다. 사실 그 이후를 알아야 더 좋은데, 아쉽습니다.

이 이후 수치형 변수로 만들어버릴 것입니다. 근로 년수로 나이를 추정할 수 있습니다. 실제 대출에서는 나이가 많으면 상환능력이 부족할 수 있다고 판단, 대출이 잘 되지 않을 것이니까요.

```python
def work_year_check(s):
    if s in ['< 1 year', '<1 year', '< 1 years']:
        return '0'
    elif s in ['10+years', '10+ years']:
        return '10'
    elif s == 'Unknown':
        return '0'
    else:
        return str(s[0])
```

```python
train_df['근로기간'] = train_df['근로기간'].apply(lambda x : int(work_year_check(x)))
test_df['근로기간'] = test_df['근로기간'].apply(lambda x : int(work_year_check(x)))
```

## 데이터 추가 전처리 및 상관관계 분석

```python
train_df
```

데이터에 대한 전처리는 다음과 같이 진행하고자 합니다.
- ID는 DROP.
- 대출금액, 연간소득, 총상환원금, 총상환이자, 연소득 등 돈 관련 금액이 폭이 크나, 그대로 사용하여 연봉이 높은 고객을 살리자.
  - 값이 크다는 그 자체로도 의미있습니다.
  - 위에서 일부 예시나 시각화 결과를 보았을 때, kNN 등의 방법론들은 적합하지 않을 것이라 예상됩니다. 너무 몰려 있습니다.
- 대출 기간은 month를 12로 나눠서 year 단위로 변환하자.
- 주택 소유 여부는 is_mortgage, is_rent, is_own으로 3가지로 구분하자. 
  - test에 대한 정보를 몰라서, 이게 가장 합리적입니다. any를 함부로 판단하지 맙시다.
- 대출 목적은 부채 통합, 신용 카드, 주택 개선과 주택 외에 모두 기타로 취급하여 통합하겟습니다.

추가로, 상관관계를 분석하기 위해 등급을 A, B, C, D, E, F, G를 6, 5, 4, 3, 2, 1, 0 으로 변환하여 상관관계를 읽겠습니다.
정말 유의미한 변수가 하나라도 제대로 존재할까요?

```python
# train_df['대출기간'].apply(lambda x: list(x.split(" ")))
```

```python
# 대출기간 맨 처음에 공백이 들어가있습니다.
train_df['대출기간'] = train_df['대출기간'].apply(lambda x: int(list(x.split(" "))[1])/12)
train_df['is_rent'] = train_df['주택소유상태'].apply(lambda x: 1 if x == 'RENT' else 0)
train_df['is_mort'] = train_df['주택소유상태'].apply(lambda x: 1 if x == 'MORTGAGE' else 0)
train_df['is_own'] = train_df['주택소유상태'].apply(lambda x: 1 if x == 'OWN' else 0)
train_df['is_부채통합'] = train_df['대출목적'].apply(lambda x: 1 if x == '부채 통합' else 0)
train_df['is_주택개선'] = train_df['대출목적'].apply(lambda x: 1 if x in ['주택 개선', '주택'] else 0)
train_df['is_신용카드'] = train_df['대출목적'].apply(lambda x: 1 if x == '신용 카드' else 0)
train_df['is_그외목적'] = train_df['대출목적'].apply(lambda x: 1 if x not in ['부채 통합', '주택 개선', '신용 카드'] else 0)

# 0-1으로 수동으로 변환시킨 변수들도 우선 제거
train_df = train_df.drop(['ID', '주택소유상태', '대출목적'], axis = 1)
```

## 추가시도 - 또 다른 변수 없을까요?

그러면, 상환 능력을 평가할 다른 변수는 또 없을까요?, 총 자산을 한번 구해보면 어떨까요?
물론, 근로연수가 고려되지 않기에, 주의해야 할 것입나다만 기대 총 자산은 구할수가 있겠죠.
그리고 미래에 벌 돈도 중요하지만, 그래서 지금 가지고 있는 돈+ 근 시일내에 벌 돈이 어느정도일지 중요하죠.
따라서 예상 총 자산을 다음과 같이 구할 수 있습니다.

- 예상 총 자산 = (근무연도 + 1) * 연간소득

0년차여도 신입사원일 수 있으니 근 시일의 소득은 고려해줍시다.

```python
X['예상총자산'] = (X['근로기간']+1) * X['연간소득']
test_df['예상총자산'] = (test_df['근로기간']+1) * test_df['연간소득']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, stratify = y, random_state = 42)
```

```python
# 위에서는 훈련 셋에다 만들었습니다. 따라서 원본 데이터에 다시 공식을 적용하여 시각화.
train_df['예상총자산'] = (train_df['근로기간']+1) * train_df['연간소득']

plt.rcParams["figure.figsize"] = (10,6)
sns.boxplot(x = '대출등급', y = '예상총자산', data = train_df,
           order = ['A', 'B', 'C', 'D', 'E', 'F', 'G'])
plt.show()
```

이상점만 보이기는 한데, 등급에 따라 차이가 있을거 같기도 합니다. 한번 학습에 반영해봅시다.

```python
model = XGBClassifier(max_depth = 7, random_state = 42, objective = 'multi:softmax', eval_metric = 'logloss',
                     learning_rate = 0.1)
model.fit(X_train, y_train)
pred = model.predict(X_test)
print(f1_score(pred, y_test, average = 'macro'))
```

```python
plt.rcParams["figure.figsize"] = (10,6)
plot_importance(model)
plt.show()
```

예상 총자산의 중요도가 상당히 높게 나왔습니다. 연간소득과 거의 같은 비중입니다. 
- 예상 총자산은 현재 가지고 있는 현금성 자산의 총액수를 추정할 수 있습니다.
- 연간소득은 단기적 기준입니다.
따라서, 둘 다 필요합니다!

하지만 여기에는 큰 문제가 있습니다. 실제 결과도 미세하게 낮게 나왔습니다.
- 10년 차 이상의 근로 연수를 가진 사람의 실제 총 자산을 알 수 없다.
- 자산이 많더라도 근로 연수가 많으면 나이가 많아 결국 대출이 안나온다는 굴레에 빠진다.
- 소득의 사용에 대해서는 고려되지 않아 부정확한 결과를 가져온다.

중요한 변수로 사용된 것으로 보이고 변수를 늘려서 성능이 좋아진 것으로 보이나, 실제 결과는 좋지 않습니다. 동일한 depth로 실험해보면 그 결과를 알 수 있고요. 단순 score 비교로도 알 수 있습니다.
Overfitting의 경우도 존재하겠으나 우선 데이터를 정확하게 설명해 줄 수 없습니다.

대신 급여 데이터 자체가 단기적인 현금 확보를 설명해주므로, 변수를 없애도 괜찮을 것입니다. 합쳐서 오히려 문제가 발생한 거니까요.
